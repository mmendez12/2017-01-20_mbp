## The Jupyter notebook through time and history

- **2001**: Fernando Pérez develops IPython
    - It is a fancy Python shell which integrates basic code editor features:
        - Syntax highlighting
        - Autocompletion
        - Introspection
        - Etc
- **2011**: Inspired by the Wolfram Mathematica notebook, the IPython notebook is officially released and shipped with IPython
- **2014**: IPython notebook becomes Jupyter (Julia, Python, R) notebook
    - The goal is to port the notebook system to other languages​
    
*Ref: https://www.datacamp.com/community/tutorials/tutorial-jupyter-notebook#gs.qMh7z3w*

## Environment and architecture

- **Notebook document**
    - Notebook documents (or “notebooks”, all lower case) are documents produced by the Jupyter Notebook App, which contain both computer code (e.g. python) and rich text elements (paragraph, equations, figures, links, etc...). Notebook documents are both human-readable documents containing the analysis description and the results (figures, tables, etc..) as well as executable documents which can be run to perform data analysis.
- **Jupyter Notebook App**
    - The Jupyter Notebook App is a server-client application that allows editing and running notebook documents via a web browser. The Jupyter Notebook App can be executed on a local desktop requiring no internet access (as described in this document) or can be installed on a remote server and accessed through the internet.
    - In addition to displaying/editing/running notebook documents, the Jupyter Notebook App has a “Dashboard” (Notebook Dashboard), a “control panel” showing local files and allowing to open notebook documents or shutting down their kernels.
- **kernel**
    - A notebook kernel is a “computational engine” that executes the code contained in a Notebook document. The ipython kernel, referenced in this guide, executes python code. Kernels for many other languages exist (official kernels).
    - When you open a Notebook document, the associated kernel is automatically launched. When the notebook is executed (either cell-by-cell or with menu Cell -> Run All), the kernel performs the computation and produces the results. Depending on the type of computations, the kernel may consume significant CPU and RAM. Note that the RAM is not released until the kernel is shut-down.
- **Notebook Dashboard**
    - The Notebook Dashboard is the component which is shown first when you launch Jupyter Notebook App. The Notebook Dashboard is mainly used to open notebook documents, and to manage the running kernels (visualize and shutdown).
    - The Notebook Dashboard has other features similar to a file manager, namely navigating folders and renaming/deleting files.

Ref: https://jupyter-notebook-beginner-guide.readthedocs.io/en/latest/what_is_jupyter.html​

## Modes 

There is two modes available.You can guess the mode you are in by looking at the color contour of the active cell:

- **Edit mode** (green contour)
    - Edit text
    - Write code
- **Command mode** (blue contour)
    - Run code
    - Use shortcuts to navigate through the notebook

## Editing Rich text

By default cells are configured to interpret code. In order to write text, you need to change the type of the cell to 'Markdown' from the toolbar by clicking on the dropdown menu available located next to the refresh button.

*Tip: you can also use the shortcut 'M' in command mode*

In Edit mode, you can write text using the Github markdown syntax. For example you can write: 
- *Italic text* using \*
- **Bold text** using \**
- [link](http://google.com) using \[text\]\(url\)
- list using \-

You can also write titles and subtitles by preceding your text with multiple \#:

# Title 1
## Title 2
### Title 3

But you can also write formulas inline with $\LaTeX$ by flanking the expression with the $ sign.

Or you can create a block on a new line using $$:

$$ P(A \mid B) = \frac{P(B \mid A) \, P(A)}{P(B)} $$

````python
l = range(1000000)

def stringify_for_loop():
    my_list = []
    for e in l:
        my_list.append(str(e))
    return my_list
        
def stringify_map():
    return map(str, l)

def stringify_list_comp():
    return [str(e) for e in l]
````

````python
%timeit stringify_for_loop()
````

````python
%timeit stringify_map()
````

````python
%timeit stringify_list_comp()
````

## NBconvert 

The notebooks can be converted to different format from the Notebook server and from the command line.

From the server, open the notebook file and click on File -> Download as to see the list of available formats. This will convert and download the notebook into your default download directory

From the command line, you can simply type `jupyter nbconvert <source.ipynb> --to [html|pdf|python ...]` to automatically convert the notebook into your current directory

## Sharing the notebooks

If you simply want to share your notebook with collaborators, your best option is to convert the notebook file to HTML and send it via email.


If you are looking for a more permanent solution or simply want to make your notebooks publicly available, you can use GitHub which now automatically renders \*.ipynb files. You can create an account for free and share your notebooks by copy pasting the content of the notebook into a Gist. However if you want to share a collection of notebooks, you must create a repository and upload all your notebooks. 

Note that the \*.ipynb files are actually simple JSON objects. This format is quite verbose and can be used efficiently by versionning tools.

Alternatively, if you can store your \*.ipynb files on a public store accessible from a static URL, you can use the link to visualize the notebooks on the [nbviewer website](http://nbviewer.jupyter.org/)

## Annoying Behavior

The output of the print function is stored in the \*ipynb file and render as HTML in your browser, limiting the amount of information that you can print. If you are trying to print too much information, your web browser will certainly crash and you will have to restart the kernel manually. Most of the time the work is saved automatically but you might loose some of your work. I highly recommend to version control your notebooks and commit/push your work as much as possible.
